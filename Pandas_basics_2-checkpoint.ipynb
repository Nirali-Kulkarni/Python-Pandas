```python
#Creating new df df1
import pandas as pd
import numpy as np   
dict1 ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv',
                'Mittal','Archana'],
                'Marks':[98,np.nan,99,87,np.nan,83,99],
                'Gender':['Male','Female','Female','Male','Male',
                         'Female','Female'],
                'Email':['priyang@gmail.com','aadhya@gmail.com',
                        'krisha@gmail.com','vedant@yahoo.com','parshv@hotmail.com',
                         'mittal@yahoo.com','archana@yahoo.com']
               }
df1=pd.DataFrame(dict1)
df1
```

```python
#Filling NA values with mean marks
df1['Marks']=df1['Marks'].fillna(df1['Marks'].mean()) 
```

```python
df1
```

```python
#sorting marks in ascending order
df1['Marks'].sort_values(ascending=False)
#Finding highest 5 Marks
df1.nlargest(5,'Marks')
#Finding student names with highest marks and sorting in ascending order
df1[df1['Marks']==df1['Marks'].max()]['Name'].sort_values()
```

```python
#creating new dataframe df2
df2=pd.DataFrame({'Name':['Nirali'],
                'Marks':[98],
                'Gender':['Female'],
                'Email':['nirali@gmail.com']
    
})
df2
```

```python
#Concatenating 2 dataframes
#without ignore_index=True next dataframe will start from 0 index
df3=pd.concat([df1,df2],ignore_index=True)
df3
```

```python
#loc and iloc
#With loc you can access column name
#with iloc you can access row number
#[:,:] indicates all the rows and all the columns
df3.loc[:,:]
#[1:5,:] indicates 1to 5 rows and all the columns
df3.loc[1:5,'Name':'Gender']
#iloc only takes number
df3.iloc[1:6,:]
df3.iloc[2:7,3]
```

```python
#creating a column with name and marks
#bit marks in integer column and cant concatenate with name which is string
#so here need to first convert marks to str and then concatenate with name
df3['name_marks']=df3['Name']+','+df3['Marks'].apply(str)
df3
```

```python
#Using group by function
#finding highest avg value per gender
df3.groupby('Gender')['Marks'].mean()
```
